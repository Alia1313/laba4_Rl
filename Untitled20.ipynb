```python
import sys
import os
import numpy as np
from collections import defaultdict
import random
import matplotlib.pyplot as plt

import gymnasium as gym

class QLearningAgent:
    def __init__(self, alpha, epsilon, discount, get_legal_actions):
        self.get_legal_actions = get_legal_actions
        self._qvalues = defaultdict(lambda: defaultdict(lambda: 0))
        self.alpha = alpha
        self.epsilon = epsilon
        self.discount = discount

    def get_qvalue(self, state, action):
        return self._qvalues[state][action]

    def set_qvalue(self, state, action, value):
        self._qvalues[state][action] = value

    def get_value(self, state):
        possible_actions = self.get_legal_actions(state)

        if len(possible_actions) == 0:
            return 0.0

        q_values = [self.get_qvalue(state, action) for action in possible_actions]
        value = max(q_values)

        return value

    def update(self, state, action, reward, next_state):
        gamma = self.discount
        learning_rate = self.alpha

        current_q = self.get_qvalue(state, action)
        next_v = self.get_value(next_state)
        target_q = reward + gamma * next_v
        new_q = (1 - learning_rate) * current_q + learning_rate * target_q

        self.set_qvalue(state, action, new_q)

    def get_best_action(self, state):
        possible_actions = self.get_legal_actions(state)
        if len(possible_actions) == 0:
            return None
        best_action = max(possible_actions, key=lambda a: self.get_qvalue(state, a))
        return best_action

    def get_action(self, state):
        possible_actions = self.get_legal_actions(state)
        if len(possible_actions) == 0:
            return None

        if random.random() < self.epsilon:
            chosen_action = random.choice(possible_actions)
        else:
            chosen_action = self.get_best_action(state)

        return chosen_action

def play_and_train(env, agent, t_max=10**4):
    total_reward = 0.0
    s, _ = env.reset()

    for t in range(t_max):
        a = agent.get_action(s)
        next_s, r, terminated, truncated, _ = env.step(a)
        done = terminated or truncated

        agent.update(s, a, r, next_s)
        s = next_s
        total_reward += r

        if done:
            break

    return total_reward

def train_taxi():

    env = gym.make("Taxi-v3")
    n_actions = env.action_space.n

    agent = QLearningAgent(
        alpha=0.5,
        epsilon=0.25,
        discount=0.99,
        get_legal_actions=lambda s: range(n_actions)
    )

    rewards = []
    for i in range(1000):
        reward = play_and_train(env, agent)
        rewards.append(reward)
        agent.epsilon *= 0.99

        if i % 100 == 0:
            print(f"Episode {i}: epsilon={agent.epsilon:.4f}, "
                  f"mean reward (last 10)={np.mean(rewards[-10:]):.2f}")

    env.close()

    plt.figure(figsize=(10, 5))
    plt.plot(rewards)
    plt.title('Reward per Episode (Taxi-v3)')
    plt.xlabel('Episode')
    plt.ylabel('Reward')
    plt.grid(True)
    plt.tight_layout()
    plt.savefig('taxi_reward_per_episode.png', dpi=150)
    plt.show()

    print(f"\nСредняя награда за последние 100 эпизодов: {np.mean(rewards[-100:]):.2f}")

    return agent, rewards


class Binarizer(gym.ObservationWrapper):
    def __init__(self, env, n_digits=None):
        super().__init__(env)
        if n_digits is None:
            n_digits = [1, 3, 2, 3]
        self.n_digits = n_digits

    def observation(self, observation):
        state_binarized = tuple(
            round(float(observation[i]), self.n_digits[i])
            if i < len(self.n_digits)
            else round(float(observation[i]), 2)
            for i in range(len(observation))
        )
        return state_binarized

def play_and_train_binarized(env, agent, t_max=1000):
    total_reward = 0.0
    s, _ = env.reset()

    for t in range(t_max):
        a = agent.get_action(s)
        next_s, r, terminated, truncated, _ = env.step(a)
        done = terminated or truncated

        agent.update(s, a, r, next_s)
        s = next_s
        total_reward += r

        if done:
            break

    return total_reward


def train_cartpole():

    env = Binarizer(gym.make("CartPole-v1"))
    n_actions = env.action_space.n
    seen_observations = []
    for _ in range(500):
        obs, _ = env.reset()
        seen_observations.append(obs)
        done = False
        while not done:
            action = env.action_space.sample()
            obs, r, terminated, truncated, _ = env.step(action)
            done = terminated or truncated
            seen_observations.append(obs)

    seen_observations = np.array(seen_observations, dtype=object)

    fig, axes = plt.subplots(1, 4, figsize=(15, 3))
    for obs_i in range(4):
        values = [obs[obs_i] for obs in seen_observations]
        axes[obs_i].hist(values, bins=20, edgecolor='black', alpha=0.7)
        axes[obs_i].set_title(f'Dimension {obs_i}')
        axes[obs_i].set_xlabel('Value')
        axes[obs_i].set_ylabel('Frequency')
        axes[obs_i].grid(True, alpha=0.3)
    plt.tight_layout()
    plt.savefig('cartpole_observations.png', dpi=150)
    plt.show()

    agent = QLearningAgent(
        alpha=0.1,
        epsilon=0.2,
        discount=0.99,
        get_legal_actions=lambda s: range(n_actions)
    )

    rewards = []
    for i in range(500):
        reward = play_and_train_binarized(env, agent)
        rewards.append(reward)
        agent.epsilon *= 0.995

        if i % 50 == 0:
            print(f"Episode {i}: epsilon={agent.epsilon:.4f}, "
                  f"mean reward (last 10)={np.mean(rewards[-10:]):.2f}")

    env.close()

    print(f"\nСредняя награда за последние 100 эпизодов: {np.mean(rewards[-100:]):.2f}")

    return agent, rewards

def test_agent(env, agent, n_episodes=100, render=False):
    old_epsilon = agent.epsilon
    agent.epsilon = 0

    total_rewards = []

    for episode in range(n_episodes):
        s, _ = env.reset()
        episode_reward = 0
        done = False

        while not done:
            if render:
                env.render()

            a = agent.get_action(s)
            next_s, r, terminated, truncated, _ = env.step(a)
            done = terminated or truncated

            s = next_s
            episode_reward += r

        total_rewards.append(episode_reward)

    agent.epsilon = old_epsilon

    return np.mean(total_rewards), np.std(total_rewards)


def main():
    taxi_agent, taxi_rewards = train_taxi()
    cartpole_agent, cartpole_rewards = train_cartpole()


    taxi_env = gym.make("Taxi-v3")
    taxi_mean, taxi_std = test_agent(taxi_env, taxi_agent, n_episodes=100)
    print(f"Taxi-v3: Средняя награда = {taxi_mean:.2f} ± {taxi_std:.2f}")
    taxi_env.close()

    cartpole_env = Binarizer(gym.make("CartPole-v1"))
    cartpole_mean, cartpole_std = test_agent(cartpole_env, cartpole_agent, n_episodes=100)
    print(f"CartPole-v1: Средняя награда = {cartpole_mean:.2f} ± {cartpole_std:.2f}")
    cartpole_env.close()


if __name__ == "__main__":
    main()
```
